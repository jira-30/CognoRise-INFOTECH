BUILDING MY MODEL USING SVM

*since I imported sklearn, Using the following*
1.   SVM--> Support Vector Machine
2.   SVC--> Support Vector Classifier



```python
# Model Training: step 1--> Splitting the features and target
data=dataset.values
X= data[:,0:4]
Y= data[:,4]
```

```python
# Model Training: step 2--> splitting dataset into train and test datset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

```python
# Model Training: step 3--> Training model using svm algorithm
from sklearn.svm import SVC
classifier=SVC()
classifier.fit(X_train,y_train)
```

FINAL PREDICTION

```python
# Model Training: step 4--> Testing model using svc, again
X_new = np.array([[5.1,3.8,1.5,0.3],[6.4,3.2,4.5,1.5],[7.2,3,5.8,1.6]])

# Prection Model
prediction= classifier.predict(X_new)
print("Prediction of Species:",format(prediction))
```
