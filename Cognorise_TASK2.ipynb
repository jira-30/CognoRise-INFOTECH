```python
# reading the dataset
Card_data=pd.read_csv("/content/creditcard.csv")
```

PRE PROCESSING

```python
from sklearn.preprocessing import StandardScaler
Card_data['normalizedAmount'] = StandardScaler().fit_transform(Card_data['Amount'].values.reshape(-1,1))
data = Card_data.drop(['Amount'],axis=1)
```

```python
data.head
```

```python
data = Card_data.drop(['Time'],axis=1)
data.tail
```

BUILDING MY MODEL USING RANDOM FOREST CLASSIFIER

*since I imported sklearn, Using the following*
1.   ensemble
2.   RandomForestClassifier



```python
# Model Training: step 1--> Splitting the features and target
X = data.iloc[:, data.columns != 'Class']
Y = data.iloc[:, data.columns == 'Class']
```

```python
X.head()
```

```python
Y.head()
```

```python
# Model Training: step 2--> splitting dataset into train and test datset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
# Model Training: step 3--> Training model using RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
ran_forest=RandomForestClassifier(n_estimators=100)
ran_forest.fit(X_train,y_train.values.ravel())
```

FINAL PREDICTION

```python
Prediction= ran_forest.predict(X_test)
```

```python
ran_forest.score(X_test,y_test)
```

VISUALIZATION MODEL:





1. using Confusion matrix
2. using seaborn HeatMap   








```python
from sklearn.metrics._plot.confusion_matrix import confusion_matrix
data_matrix=confusion_matrix(y_test,Prediction)
```

```python
labels = [0,1]
sns.heatmap(data_matrix, annot=True, cmap='YlGnBu', fmt=".3f", xticklabels=labels, yticklabels=labels)
plt.show()
```
